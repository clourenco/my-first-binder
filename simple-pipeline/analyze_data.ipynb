```python
# analyze_data.ipynb

# Import necessary libraries
import pandas as pd

# Load the preprocessed data
data = pd.read_csv('preprocessed_data.csv')

# Perform a simple analysis (e.g., calculate the average salary)
average_salary = data['salary'].mean()

# Save the analysis result
with open('analysis_result.txt', 'w') as f:
    f.write(f"Average salary of people aged 30 or below: {average_salary}\n")

print("Analysis complete. Result saved as 'analysis_result.txt'")
```
