```python
# preprocess_data.ipynb

# Import necessary libraries
import pandas as pd

# Load a sample dataset (you can replace this with your dataset)
data = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [24, 27, 22, 32, 29],
    'salary': [50000, 60000, 55000, 65000, 70000]
})

# Preprocess the data (for example, filter out people older than 30)
preprocessed_data = data[data['age'] <= 30]

# Save the preprocessed data to a CSV file for the next step
preprocessed_data.to_csv('preprocessed_data.csv', index=False)

print("Data preprocessed and saved as 'preprocessed_data.csv'")
```
